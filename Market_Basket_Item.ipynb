***Market Basket Item - Apriori Algorithm Dataset from real-world retailer***

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Loading the Data
data = pd.read_csv('/content/TRAIN-ARULES.csv')
data.head()
```

```python
# Exploring the columns of the data
data.columns
```

```python
# Grouping data
data=data.groupby(['order_id', 'product_name'])['user_id'].sum().unstack().reset_index().fillna(0).set_index('order_id')
data
```

```python
# Defining the hot encoding function to make the data suitable for the libraries
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
```

```python
# Encoding the dataset
data_encoded = data.applymap(hot_encode)
data_encoded
```

```python
# Building the model
frq_items = apriori(data_encoded, min_support = 0.0045, use_colnames = True)
frq_items
```

```python
# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric ="support", min_threshold = 0.0045).sort_values('support', ascending=False).reset_index(drop=True)
rules
```

```python
# Predicting next items in the basket based for given testing data
test_dataset = rules[rules['antecedents'] == {'Dark Chocolate Minis', 'Organic Pink Lemonade Bunny Fruit Snacks', 'Peach-Pear Sparkling Water'}]
```
